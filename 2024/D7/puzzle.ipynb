```python
with open("input.txt") as inFile:
    equations = [line.strip().split(":") for line in inFile.readlines()]
```

```python
def getResult(nums):
    if len(nums) == 1:
        return [nums[0],]
    
    subResults = getResult(nums[1:])

    results = []
    for subResult in subResults:
        results.append(nums[0] + subResult)
        results.append(nums[0] * subResult)

    return results
```

Part 1

```python
trueEquations = []

for equation in equations:
    testValue, nums = equation
    testValue = int(testValue)
    nums = list(reversed([int(num) for num in nums.split()]))

    if testValue in getResult(nums):
        trueEquations.append(testValue)

sum(trueEquations)
```

```python
def getResultPart2(nums):
    if len(nums) == 1:
        return [nums[0],]
    
    subResults = getResultPart2(nums[1:])
    
    results = []
    for subResult in subResults:
        results.append(nums[0] + subResult)
        results.append(nums[0] * subResult)
        results.append(int(str(subResult) + str(nums[0])))
    
    return results
```

```python
trueEquations = []

for equation in equations:
    testValue, nums = equation
    testValue = int(testValue)
    nums = list(reversed([int(num) for num in nums.split()]))
    
    results = getResultPart2(nums)

    # print()
    # print("Test:", testValue, "Nums:", nums)
    # print("Result:", results)

    if testValue in results:
        trueEquations.append(testValue)

# print(trueEquations)
sum(trueEquations)
```
